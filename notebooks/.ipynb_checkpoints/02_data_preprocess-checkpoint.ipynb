```python
import pandas as pd
import os
import numpy as np
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
import re
```

```python
df = pd.read_csv("C:/Users/yunus/Desktop/Bahar Dersleri/Yapay Zeka Proje/arac-arizadeseni-eslestirme/data/raw/all_data.csv")
```

```python
df.head()
```

```python
def preprocess_text(text):
    #Küçük harfe çevirme
    text = text.lower()
    #Noktalama İşaretlerini kaldırma
    text = re.sub(r'[^\w\s]', '', text)
    # Tokenizasyon
    tokens = nltk.word_tokenize(text)
    # Stop word
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]

    #Lemmatizasyon
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return tokens
```

```python
# Fonksiyonu kullanarak işlemleri uygulama
df['comments'] = df['comments'].apply(preprocess_text)
```

```python
def stem_text(text):
  # Küçük harfe çevirme
  text = text.lower()
  # Noktalama işaretlerini kaldırma
  text = re.sub(r'[^\w\s]', '', text)
  # Tokenizasyon
  tokens = nltk.word_tokenize(text)
  # Stop word temizleme
  stop_words = set(stopwords.words('english'))
  tokens = [token for token in tokens if token not in stop_words]
 

  # Stemming
  stemmer = PorterStemmer()
  tokens = [stemmer.stem(token) for token in tokens]
  return tokens
```

```python
df2['comments_stemmed'] = df2['comments'].apply(stem_text)
```
